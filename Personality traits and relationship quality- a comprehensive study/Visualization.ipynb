```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr


data = pd.read_excel('combined_score.xlsx')  
traits = ['Extraversion_score', 'Agreeableness_score', 'Conscientiousness_score', 'Neuroticism_score', 'Openness_score']

# Create scatter plots
plt.figure(figsize=(15, 10))
plt.rcParams['font.family'] = 'Arial'
for i, trait in enumerate(traits, 1):
    plt.subplot(2, 3, i)  # Adjust layout for 5 plots
    plt.scatter(data[trait], data['CSI_Total_Score'], color='skyblue', edgecolor='gray', alpha=0.7)
    plt.title(f'{trait} vs CSI Total Score', fontweight='bold', fontsize=10)
    plt.xlabel(f'{trait}', fontsize=10)
    plt.ylabel('CSI Total Score', fontsize=10)
    plt.grid(alpha=0.4)

plt.tight_layout(pad = 4)
plt.savefig('scatter_plots.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
