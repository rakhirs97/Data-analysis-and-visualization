```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Step 1: Read the Excel file
data = pd.read_excel('combined_score.xlsx')  


# Step 2: Select columns of interest for correlation analysis
columns_of_interest = ['Extraversion_score', 'Agreeableness_score', 'Conscientiousness_score', 
                        'Neuroticism_score', 'Openness_score', 'CSI_Total_Score']
data_subset = data[columns_of_interest]

# Step 3: Perform Pearson Correlation
correlation_matrix = data_subset.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix between Personality Traits and Relationship Quality")
plt.show()

from scipy.stats import spearmanr
correlation, p_value = spearmanr(data['Extraversion_score'], data['CSI_Total_Score'])
print(f"Spearman correlation: {correlation}, p-value: {p_value}")

# Optional: For specific pairwise correlation, use Pearson's r test
# Example: Correlation between Extraversion and CSI Total Score
corr, p_value = pearsonr(data['Extraversion_score'], data['CSI_Total_Score'])
print(f"Pearson correlation between Extraversion and CSI Total Score: {corr}, p-value: {p_value}")

plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(data['Extraversion_score'], data['CSI_Total_Score'], color='blue', alpha=0.7)  # Scatter plot
plt.title('Scatter Plot: Extraversion vs CSI Total Score')  # Title
plt.xlabel('Extraversion Score')  # x-axis label
plt.ylabel('CSI Total Score')  # y-axis label
plt.grid(True)  # Optionally add grid lines
plt.show()  # Display the plot
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import spearmanr

# Step 1: Read the Excel file
data = pd.read_excel('combined_score.xlsx') 

# Calculate Spearman's rank correlation for Extraversion
spearman_extraversion, p_extraversion = spearmanr(data['Extraversion_score'], data['CSI_Total_Score'])

# Calculate Spearman's rank correlation for Agreeableness
spearman_agreeableness, p_agreeableness = spearmanr(data['Agreeableness_score'], data['CSI_Total_Score'])

# Calculate Spearman's rank correlation for Agreeableness
spearman_neuroticism, p_neuroticism = spearmanr(data['Neuroticism_score'], data['CSI_Total_Score'])

# Optional: For specific pairwise correlation, use Pearson's r test

Pearson_Openness, p_openness = pearsonr(data['Openness_score'], data['CSI_Total_Score'])

Pearson_Conscientiousness, p_Conscientiousness = pearsonr(data['Conscientiousness_score'], data['CSI_Total_Score'])

# Display the results
print("Spearman's Rank Correlation for Extraversion and CSI Total Score:")
print(f"Correlation coefficient: {spearman_extraversion:.3f}, p-value: {p_extraversion:.3f}")

print("\nSpearman's Rank Correlation for Agreeableness and CSI Total Score:")
print(f"Correlation coefficient: {spearman_agreeableness:.3f}, p-value: {p_agreeableness:.3f}")

print("\nSpearman's Rank Correlation for neuroticism and CSI Total Score:")
print(f"Correlation coefficient: {spearman_neuroticism:.3f}, p-value: {p_neuroticism:.3f}")

print("\nPearson's Rank Correlation for Openness and CSI Total Score:")
print(f"Correlation coefficient: {Pearson_Openness:.3f}, p-value: {p_openness:.3f}")

print("\nPearson's Rank Correlation for Conscientiousness and CSI Total Score:")
print(f"Correlation coefficient: {Pearson_Conscientiousness:.3f}, p-value: {p_Conscientiousness:.3f}")
```
