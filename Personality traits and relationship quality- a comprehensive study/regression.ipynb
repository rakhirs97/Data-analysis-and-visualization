```python
import pandas as pd
import statsmodels.api as sm

# Load the data
file_path = "combined_score.xlsx"  
data = pd.read_excel(file_path)

# Define dependent and independent variables
dependent_variable = 'CSI_Total_Score'
independent_variables = ['Extraversion_score', 'Agreeableness_score', 
                         'Conscientiousness_score', 'Neuroticism_score', 
                         'Openness_score']

# Prepare the data
X = data[independent_variables]  # Independent variables
y = data[dependent_variable]     # Dependent variable

# Add a constant term to the independent variables (for intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
file_path = "combined_score.xlsx"  
data = pd.read_excel(file_path)

# Fit the regression model
X = data[['Extraversion_score', 'Agreeableness_score', 'Conscientiousness_score', 'Neuroticism_score', 'Openness_score']]
X = sm.add_constant(X)
y = data['CSI_Total_Score']
model = sm.OLS(y, X).fit()

# Generate partial regression plots
fig = plt.figure(figsize=(15, 10))
sm.graphics.plot_partregress_grid(model, fig=fig)
plt.rcParams['font.family'] = 'Arial'
# Customize the plot
plt.suptitle('Partial Regression Plots', fontsize=12, fontweight='bold')
plt.subplots_adjust(top=0.9, wspace=0.3, hspace=0.3)  # Adjust spacing for readability

plt.savefig('regression.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
